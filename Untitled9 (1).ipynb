**Exercice 1**

**Suppose that an analog signal is given as: x(t) = 5 cos(2π 1000 t), for t≥0 and is sampled at the rate of 8 kHz.**

**a. Sketch the spectrum for the original signal.**

```python
import matplotlib.pyplot as plt

plt.figure()
plt.stem([1000, -1000], [2.5, 2.5])
plt.xlabel("Fréquence (Hz)")
plt.ylabel("Amplitude")
plt.title("Spectre du signal analogique")
plt.grid(True)
plt.show()
```

**b. Sketch the spectrum for the sampled signal from 0 to 20 kHz.**

```python
import matplotlib.pyplot as plt

Fs = 8000
f0 = 1000

freqs_pos = [f0 + k*Fs for k in range(-3, 4)]
freqs_neg = [-f0 + k*Fs for k in range(-3, 4)]

freqs = freqs_pos + freqs_neg
amps = [2.5]*len(freqs)

plt.figure()
plt.stem(freqs, amps)
plt.xlim(-20000, 20000)
plt.xlabel("Fréquence (Hz)")
plt.ylabel("Amplitude")
plt.title("Spectre du signal échantillonné (bilatéral)")
plt.grid(True)
plt.show()
```

**Synthétique et Technique** 

Le signal $x(t)$ étant purement sinusoïdal à $1000$ Hz, la condition de Nyquist-Shannon est largement respectée ($F_e = 8$ kHz $> 2f_{max}$). En conséquence, le signal ne subit aucun repliement spectral (aliasing)11. Le spectre résultant se caractérise par une périodicité de $8$ kHz, où le motif spectral original se répète de manière identique autour de chaque multiple de la fréquence d'échantillonnage2.

**Exercice 2**

**2. Assuming that an analog signal is given by: x(t) = 5 cos(2π 2000 t) + 3 cos(2π 3000 t), for t ≥0 and it is
sampled at the rate of 8000 Hz**

**a. Sketch the spectrum of the sampled signal up to 20 kHz.**

```python
import matplotlib.pyplot as plt

Fs = 8000
freqs = []
amps = []

for f, a in zip([2000, 3000], [2.5, 1.5]):
    for k in range(-3, 4):
        # +f autour de kFs
        freqs.append(k*Fs + f)
        amps.append(a)

        # -f autour de kFs (symétrie)
        freqs.append(k*Fs - f)
        amps.append(a)

plt.figure()
plt.stem(freqs, amps)
plt.xlim(-20000, 20000)
plt.xlabel("Fréquence (Hz)")
plt.ylabel("Amplitude")
plt.title("Spectre échantillonné (symétrique)")
plt.grid(True)
plt.show()
```

**b. Sketch the recovered analog signal spectrum if an ideal LowPass Filter (LPFi) with
a cutoff frequency of 4 kHz is used to filter the sampled signal to recover the original signal.**

```python
import matplotlib.pyplot as plt

Fc = 4000   # fréquence de coupure du LPF

freqs = []
amps = []

# composantes originales du signal
for f, a in zip([2000, 3000], [2.5, 1.5]):
    # +f
    freqs.append(f)
    amps.append(a)

    # -f (symétrie)
    freqs.append(-f)
    amps.append(a)

plt.figure()
plt.stem(freqs, amps)
plt.xlim(-Fc, Fc)
plt.xlabel("Fréquence (Hz)")
plt.ylabel("Amplitude")
plt.title("Spectre du signal récupéré après LPF idéal (Fc = 4 kHz)")
plt.grid(True)
plt.show()
```

**Synthétique et Technique** 

Dans ce scénario, le signal est composé de deux fréquences distinctes : $f_1 = 2000$ Hz et $f_2 = 3000$ Hz4. La fréquence maximale du signal ($f_{max}$) est donc de $3000$ Hz5. Comme la condition de Nyquist est satisfaite ($F_e = 8000$ Hz $> 2 \times 3000$ Hz), l'échantillonnage s'effectue sans perte d'information et sans phénomène d'aliasing.
Le spectre du signal échantillonné se manifeste par une répétition périodique du spectre de base tous les $8$ kHz7. On observe ainsi les composantes originales à $\pm 2$ kHz et $\pm 3$ kHz, ainsi que leurs répliques centrées sur les multiples de la fréquence d'échantillonnage (par exemple à $8 \pm 2$ kHz et $8 \pm 3$ kHz).

**Exercice 3**

**Given an analog signal x(t) = 5 cos(2π 2000t) + cos(2π 5000t), for t≥0, which is sampled at a rate of 8000 Hz**

**a. Sketch the spectrum of the sampled signal up to 20 kHz.**

```python
import matplotlib.pyplot as plt

Fs = 8000
freqs = []
amps = []

# composantes originales
for f, a in zip([2000, 5000], [2.5, 0.5]):
    for k in range(-3, 4):
        # +f et -f
        freqs.append(k*Fs + f)
        amps.append(a)

        freqs.append(k*Fs - f)
        amps.append(a)

# ---- aliasing dû à 5000 Hz ----
f_alias = abs(Fs - 5000)   # = 3000 Hz

for k in range(-3, 4):
    freqs.append(k*Fs + f_alias)
    amps.append(0.5)

    freqs.append(k*Fs - f_alias)
    amps.append(0.5)

plt.figure()
plt.stem(freqs, amps)
plt.xlim(-20000, 20000)
plt.xlabel("Fréquence (Hz)")
plt.ylabel("Amplitude")
plt.title("Spectre échantillonné avec aliasing")
plt.grid(True)
plt.show()
```

**b. Sketch the recovered analog signal spectrum if an ideal lowpass filter with a cutoff frequency of 4 kHz is
used to recover the original signal [y(n)=x(n) in this case].**

```python
import matplotlib.pyplot as plt

Fc = 4000  # fréquence de coupure du LPF

freqs = []
amps = []

# composante originale à 2000 Hz
for f, a in zip([2000], [2.5]):
    freqs.append(f)
    amps.append(a)
    freqs.append(-f)
    amps.append(a)

# composante aliasée de 5000 Hz → 3000 Hz
for f, a in zip([3000], [0.5]):
    freqs.append(f)
    amps.append(a)
    freqs.append(-f)
    amps.append(a)

plt.figure()
plt.stem(freqs, amps)
plt.xlim(-Fc, Fc)
plt.xlabel("Fréquence (Hz)")
plt.ylabel("Amplitude")
plt.title("Spectre du signal récupéré après LPF idéal (Fc = 4 kHz)")
plt.grid(True)
plt.show()
```

**Synthétique et Technique** 


Ce signal comporte deux fréquences : $f_1 = 2000$ Hz et $f_2 = 5000$ Hz. Ici, la condition de Nyquist-Shannon n'est pas respectée pour la seconde composante, car $f_{max} = 5000$ Hz est supérieure à la fréquence de Nyquist ($F_e/2 = 4000$ Hz).Il en résulte un phénomène d'aliasing : la composante à $5$ kHz se replie à l'intérieur de la bande de base à une fréquence apparente de $|5000 - 8000| = 3000$ Hz. Le spectre du signal échantillonné contient donc des pics à $\pm 2$ kHz (signal original) et $\pm 3$ kHz (alias du signal à 5 kHz). Lors de la reconstruction avec un filtre passe-bas de $4$ kHz, le signal récupéré sera différent du signal d'origine car il inclura cette fréquence parasite de $3$ kHz.

**b. Sketch the unilateral spectrum of the sampled signal up to 40 kHz.**

```python
import numpy as np
import matplotlib.pyplot as plt

# Paramètres
Fs = 8000
f_limit = 40000

# 1. Définition des fréquences alternatives (AC) uniquement
# Elles vont se répéter tous les Fs
def X_analog_periodic(f):
    f_abs = abs(f)
    if f_abs == 2000: return 0.5       # A/2
    if f_abs == 4000: return 1.0       # A/2
    if f_abs == 6000: return 1.5       # A/2
    return 0.0

# 2. Axe des fréquences de -40k à +40k
freqs_axe = np.arange(-f_limit, f_limit + 1, 2000)
amps_result = []

# 3. Calcul du spectre
for f in freqs_axe:
    # A. Calcul de la partie périodique (Aliasing des composantes 2k, 4k, 6k)
    somme_intermediaire = 0
    for k in range(-12, 13):
        somme_intermediaire += X_analog_periodic(f - k * Fs)
    
    # B. Ajout de la composante 0.5 UNIQUEMENT à l'origine (0 Hz)
    if f == 0:
        somme_intermediaire += 0.5
        
    amps_result.append(somme_intermediaire)

# --- TRACÉ DU GRAPHIQUE ---
plt.figure(figsize=(16, 8))

# On s'assure que les zéros sont propres
amps_result = [a if a > 0.1 else 0 for a in amps_result]

# Dessin des pics
markerline, stemlines, baseline = plt.stem(freqs_axe, amps_result, 
                                          linefmt='b-', markerfmt='bo', basefmt='k-')

# Configuration du repère centré sur 0
plt.title("Question 4.c : Spectre BILATÉRAL (DC 0.5 unique à l'origine)", fontsize=14)
plt.xlabel("Fréquence (Hz)", fontsize=12)
plt.ylabel("Amplitude", fontsize=12)

# Définition des limites du repère
plt.xlim(-f_limit - 1000, f_limit + 1000)
plt.ylim(0, 3)

# Graduation détaillée tous les 2 kHz
plt.xticks(np.arange(-f_limit, f_limit + 1, 2000), rotation=90)
plt.grid(True, linestyle=':', alpha=0.5)

# Annotation des amplitudes
for f, a in zip(freqs_axe, amps_result):
    if a > 0:
        # Affiche 0.5 pour le zéro, et l'entier pour les autres
        label = f'{a}' if f == 0 else f'{int(a)}'
        plt.text(f, a + 0.1, label, ha='center', fontsize=9, fontweight='bold')

# Marquage des multiples de Fs
for k in range(-5, 6):
    plt.axvline(x=k*Fs, color='red', linestyle='--', alpha=0.2)
    plt.text(k*Fs, -0.25, f'{k}Fs', color='red', ha='center', weight='bold')



plt.tight_layout()
plt.show()
```

**d. Conclude about b and c.**

On distingue deux méthodes de visualisation pour l'analyse spectrale : l'une est cumulative (unilatérale), limitant le champ visuel aux fréquences positives, tandis que l'autre est symétrique (bilatérale). Cette seconde approche permet de mieux appréhender le concept de repliement et la répétition du motif spectral de part et d'autre de l'origine.

**f. Sketch the recovered analog signal spectrum if an ideal Lowpass filter with a cutoff frequency of 6 kHz is
used to recover the original signal [y(n)=x(n) in this case].**

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Données du Problème ---
Fs = 8000  # Fréquence d'échantillonnage (Hz)
Fc = 6000  # Fréquence de coupure du filtre (Hz)
Fc_kHz = Fc / 1000.0

# --- 1. Définir le Spectre Échantillonné Bilatéral ---
extended_freqs_kHz = np.array([-8, -6, -4, -2, 0, 2, 4, 6, 8])
# Amplitudes : [f=-8, -6, -4, -2, 0, 2, 4, 6, 8]
X_s_amplitudes = np.array([1, 2, 2, 2, 0.5, 2, 2, 2, 1]) # Note: DC à 0.5 selon votre consigne précédente

# --- 2. Application du Filtre Passe-Bas Idéal (LPFi) ---
Y_amplitudes = np.copy(X_s_amplitudes)
for i, f_kHz in enumerate(extended_freqs_kHz):
    if abs(f_kHz) > Fc_kHz:
        Y_amplitudes[i] = 0

# --- 3. Tracé du Spectre Récupéré ---
plt.figure(figsize=(12, 7), facecolor='#f8f9fa')

# Zone ombrée pour illustrer la zone de passage du filtre
plt.axvspan(-Fc_kHz, Fc_kHz, color='teal', alpha=0.1, label='Bande passante du filtre')

# Tracé des impulsions (Couleur Orange Corail)
markerline, stemlines, baseline = plt.stem(extended_freqs_kHz, Y_amplitudes, 
                                          linefmt='#FF5733', markerfmt='o', 
                                          basefmt=' ', label='Composantes filtrées Y(f)')
plt.setp(markerline, color='#FF5733', markersize=8)
plt.setp(stemlines, linewidth=2)

# Ligne de base personnalisée
plt.axhline(0, color='black', linewidth=1)

# Fréquence de coupure (Lignes verticales)
plt.axvline(Fc_kHz, color='#2C3E50', linestyle=':', linewidth=2, label=f'Coupure Fc = {Fc_kHz} kHz')
plt.axvline(-Fc_kHz, color='#2C3E50', linestyle=':', linewidth=2)

# Mise en forme du graphique
plt.title(f"Visualisation de la Reconstruction : Spectre de Sortie Y(f)\n(Filtrage Passe-bas à {Fc_kHz} kHz)", 
          fontsize=14, fontweight='bold', pad=20)
plt.xlabel("Domaine Fréquentiel (kHz)", fontsize=11)
plt.ylabel("Magnitude Bilatérale", fontsize=11)

plt.grid(True, axis='both', linestyle='--', alpha=0.4)
plt.xticks(extended_freqs_kHz)
plt.yticks(np.arange(0, 3, 0.5))
plt.xlim(-9.5, 9.5)
plt.ylim(-0.2, 2.5)

plt.legend(loc='upper right', frameon=True, shadow=True)



plt.tight_layout()
plt.show()

# --- Affichage console ---
print(f"--- Résumé du filtrage (Fc = {Fc_kHz} kHz) ---")
for f, A in zip(extended_freqs_kHz, Y_amplitudes):
    status = "Conservé" if A > 0 else "Éliminé"
    print(f"Fréquence: {f:2} kHz | Amplitude: {A:3} | État: {status}")
```

Pour isoler les fréquences du signal source, nous avons défini un gabarit de filtrage avec une bande passante s'étendant jusqu'à $6$ kHz. Toutefois, il est crucial de noter que le filtrage intervient après l'échantillonnage : si le critère de Shannon ($F_e > 2f_{max}$) n'est pas respecté en amont, le filtre laissera passer les fréquences aliassées situées dans sa bande passante.
